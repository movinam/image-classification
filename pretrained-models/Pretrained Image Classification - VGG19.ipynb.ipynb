### Imports

```
import numpy as np
import tensorflow as tf

from tensorflow.keras.preprocessing.image import img_to_array, load_img
from tensorflow.keras.applications.vgg19 import VGG19, preprocess_input, decode_predictions

from IPython.core.display import display
```

### Constants

```
FILE1 = '08 Doorway.jpg'
FILE2 = '09 Ice Cream.jpg'
FILE3 = '07 Stairs.jpg'
```

### Preprocessing Images

```
def format_image(filename):
  pic = load_img(filename, target_size=(224,224))
  pic_array = np.expand_dims(img_to_array(pic), axis=0)
  return preprocess_input(pic_array)
```

### Load VGG19

```
%%time

vgg_model = VGG19(weights='imagenet')
vgg_model.graph = tf.compat.v1.get_default_graph()
```

### Making Predictions

```
data = format_image(FILE1)
prediction = vgg_model.predict(data)
decode_predictions(prediction)
```
